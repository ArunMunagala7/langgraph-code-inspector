```python
def calculate_statistics(numbers):
    """Calculate basic statistics from a list of numbers."""
    total = sum(numbers)
    count = len(numbers)
    mean = total / count
    return {'mean': mean, 'total': total, 'count': count}
```

```python
def find_outliers(data, threshold=2):
    """Find outliers in dataset."""
    stats = calculate_statistics(data)
    mean = stats['mean']
    outliers = []
    for value in data:
        if abs(value - mean) > threshold:
            outliers.append(value)
    return outliers
```
